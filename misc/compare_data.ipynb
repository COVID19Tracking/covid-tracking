# Data comparison

This notebook compares the current google spreadhseet with Danny's coronavirusapi.com website

```python
# download coronavirusapi
df = pd.read_html("http://coronavirusapi.com/")
```

```python
data_api = df[0]
data_api.head()
```

```python
data_api = data_api.loc[:,['Name','Tested','Positive']]
```

```python
# download google sheet
df2 = pd.read_html("https://docs.google.com/spreadsheets/d/18oVRrHj3c183mHmq3m89_163yuYltLNlOmPerQ18E8w/htmlview?usp=sharing&sle=true")
```

```python
sheet = df2[0]
sheet.head()
```

```python
sheet.columns = sheet.loc[0,:]
sheet = sheet.loc[2:,['State','Positive','Negative','Pending','Total']]
```

```python
sheet.head()
```

```python
sheet.columns = [ s + '_s' for s in sheet.columns]
data_api.columns = [ s + '_a' for s in data_api.columns]

sheet = sheet.set_index('State_s', drop=True)
data_api = data_api.set_index('Name_a', drop=True)
```

```python

merge = sheet.merge(data_api, left_index=True, right_index=True)
merge.head()
```

```python
merge.Positive_s = merge.Positive_s.astype(float)
```

```python
merge.Total_s = merge.Total_s.astype(float)
```

```python
merge.query('Positive_s != Positive_a')[['Positive_s','Positive_a']]
```

```python
sorted(merge.query('Positive_s != Positive_a').index)
```

```python
merge.query('Total_s != Tested_a')[['Total_s','Tested_a']]
```

```python
sorted(merge.query('Total_s != Tested_a').index)
```
